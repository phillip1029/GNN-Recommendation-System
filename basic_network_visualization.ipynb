```python
occurrences = {}
filtered_ratings_df = ratings_df[ratings_df['book_rating'] > 0]
for book in filtered_ratings_df['isbn']:
    if book in occurrences:
        occurrences[book] += 1
    else:
        occurrences[book] = 1
sorted_occurrences = sorted(occurrences.items(), key=lambda x: x[1], reverse=True)
top_5_books = [x[0] for x in sorted_occurrences]
for book in top_5_books:
    f=open('list.txt', 'a')
    f.write(f"{book}:{occurrences[book]}\n")
    f.close()
```
